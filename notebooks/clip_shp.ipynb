# SHP Clip

```python
import geopandas as gpd
from geopandas import GeoDataFrame
```

```python
road_rail_gdf: GeoDataFrame = gpd.read_file("../data/fr/merged/route_rail.shp")
mask_gpd: GeoDataFrame = gpd.read_file("../data/fr/clip/route_rail_clip_mask.shp")
```

```python
road_rail_clipped_gdf: GeoDataFrame = gpd.clip(road_rail_gdf, mask_gpd)
road_rail_clipped_gdf.to_file("../data/fr/clip/route_rail_clip.shp")
```
