# Exemple de séparation de shapefile

```python
import geopandas as gpd
from geopandas import GeoDataFrame
from shapely.geometry import LineString, MultiLineString
from shapely import Polygon
from shapely.ops import split
from pandas import DataFrame

DATA: str = "../sat-data/"
OUT: str = "../out/"
```

```python
gdf: GeoDataFrame = gpd.read_file(f"{DATA}/2026-07-16-00_00_2026-07-16-23_59_Sentinel-2_L2A_True_color-part-1.shp")
polygon_to_split = gdf.geometry.iloc[0]
```

```python
bounds = gdf.bounds.iloc[0]
print(f"Shapefile bounds: {bounds.head()}")

middle_point = (bounds.maxx + bounds.minx) / 2
cut_line: LineString = LineString([(middle_point, bounds.miny), (middle_point, bounds.maxy)])
print(f"Shapefile will be cut at: {cut_line.bounds}")
```

```python
split_result = split(polygon_to_split, cut_line)
split_shapefiles = [geom for geom in split_result.geoms if geom.geom_type == 'Polygon']
```

```python
final_gdf: GeoDataFrame = GeoDataFrame(geometry=split_shapefiles, crs=gdf.crs)
final_gdf.to_file(f"{OUT}split_shapefile.shp")
```
