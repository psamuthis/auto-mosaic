# Exemple de fusion de shapefile

```python
import subprocess

DATA_PATH: str = "../data/"
ROAD_PATH: str = f"fr/route/TRONCON_ROUTE.shp"
RAIL_PATH: str = f"fr/rail/TRONCON_VOIE_FERREE.shp"
OUTPUT_FILE: str = f"fr/merged/route_rail.shp"
```

## Fusion d'un shp des routes avec une shp des rivières

```python
result = subprocess.run(f"""
. ../set_env.sh && 
docker run --rm -v ../data/:/data \\
$GDAL_IMAGE \\
ogrmerge.py -single -overwrite_ds -o /data/{OUTPUT_FILE} \\
/data/{ROAD_PATH} /data/{RAIL_PATH}
""", shell=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
```
