```python
import geopandas as gpd
import subprocess
from geopandas import GeoDataFrame
```

```python
AOI_SHP: str = "../data/fr/clip/route_rail_clip.shp"
OUT_SHP: str = "../data/fr/mosaic/route_rail.shp"
```

```python
aoi_gdf: GeoDataFrame = gpd.read_file(AOI_SHP)
```

```python
result = subprocess.run(f"""
docker run --rm -e QT_QPA_PLATFORM=offscreen -v ../data/:/data \\
qgis/qgis /usr/bin/qgis_process run native:shortestpathpointtopoint \\
    --verbose \\
    --no-python \\
    --skip-loading-plugins \\
    --INPUT="/data/fr/merged/route_rail.shp" \\
    --STRATEGY=1 \\
    --START_POINT="758442.116998,6735097.427300" \\
    --END_POINT="757486.482913,6778425.194135" \\
    --OUTPUT="/data/fr/mosaic/path.shp" \\
    --PROJECT_PATH="/data/qgis_project.qgz" \\
    --DEFAULT_DIRECTION=2 \\
    --DEFAULT_SPEED=50 \\
    --TOLERANCE=0
""", shell=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
```
